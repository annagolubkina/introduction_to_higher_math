#### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,	
эллипса,

```python
#окружность
plt.figure(figsize=(5, 5))

R = 20
x = np.linspace(-R, R, 200)
y1 = np.sqrt(R ** 2 - x ** 2)
y2 = -np.sqrt(R ** 2 - x ** 2)
plt.plot(x, y1, linewidth = 2, color = 'red')
plt.plot(x, y2, linewidth = 2, color = 'red')

plt.xlabel("x")
plt.ylabel("y")
```

```python
#эллипс
plt.figure(figsize = (7, 4))

a = 6
b = 2
x = np.linspace(-a, a, 1000)
y1 =  np.sqrt( b**2 * (a**2 - x**2) / a**2 )
y2 = -np.sqrt( b**2 * (a**2 - x**2) / a**2 ) 

plt.plot(x, y1, linewidth = 2, color = 'blue')
plt.plot(x, y2, linewidth = 2, color = 'blue')
    
plt.xlabel("x")
plt.ylabel("y")
```

```python
#гипербола
plt.figure(figsize = (8, 4))

a = 6
b = 2
x = np.linspace(-30, 30, 1000)
y1 =   (b / a) * np.sqrt( x**2 - a**2 )
y2 =  -(b / a) * np.sqrt( x**2 - a**2 )

plt.plot(x, y1, linewidth = 2,  color = 'green')
plt.plot(x, y2, linewidth = 2,  color = 'green')

plt.xlabel("x")
plt.ylabel("y")
```

#### 4. Решить систему уравнений и неравенств
1)
exp(x) + x∙(1 – y) = 1  
y = x2 – 1

2)
y = x2 – 1
exp(x) + x∙(1 – y)  - 1 > 0

```python
#1)
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (, 10))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()


def equation(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equation, (2, 4))
x2, y2 =  fsolve(equation, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
#2)
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
